Unenployment rate by country 1991-2021

```python
# Read the CSV file into a `pandas` DataFrame - Unemployment
unemployment_file = "https://raw.githubusercontent.com/kkruel8100/Crime-Analysis/main/resources/unemployment%20analysis.csv"
unemployment_df = pd.read_csv(unemployment_file, low_memory=False)

# Create a copy of the df
unemployment_copy = unemployment_df.copy()

# Clean the columns
cleaned_unemployment_copy = clean_columns(unemployment_copy)

# Display df head and info
display(cleaned_unemployment_copy.head())
display(cleaned_unemployment_copy.info())
```

```python
# Sort values and display tail
cleaned_unemployment_copy.sort_values("country_name", inplace=True)
cleaned_unemployment_copy.tail(20)
```

```python
# Create US unemployment df and display
unemployment_us = cleaned_unemployment_copy.loc[cleaned_unemployment_copy["country_name"] == "United States"]
unemployment_us
```

```python
# Drop US unemployment columns and display head
unemployment_us = unemployment_us.drop(columns=["country_name", "country_code"])
unemployment_us.head()
```

```python
# Create US unemployment melt
unemployment_us_melt = unemployment_us.melt()

# Rename columns
unemployment_us_melt.rename(columns={"variable": "year", "value": "unemployment_rate"}, inplace=True)

# Display head and tail
display(unemployment_us_melt.head())
display(unemployment_us_melt.tail())
```

```python
# Create US unemployment plot
unemployment_us_plot = unemployment_us_melt.hvplot(
    x="year",
    y="unemployment_rate",
    title="US Unemployment Rate",
    xlabel="Year",
    ylabel="Unemployment Rate",
    rot=45
)

# Display plot
unemployment_us_plot
```
